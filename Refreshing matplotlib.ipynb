### Matplotlib example workflow
- We want:
  - Line plot
  - Scatter plot
  - Bar plot
  - Histogram

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline 
import matplotlib.pyplot as plt

# 1. Prepare data
a = [1, 2, 3, 4]
b = [11, 22, 33, 44]

# 2. Setup plot
ig, ax = plt.subplots(figsize = (10, 10)) # (width, height)

# 3. Plot data
ax.plot(a,b)

# 4. Customize the plot
ax.set(title = 'simple plot',
       xlabel = 'X-axis',
       ylabel = 'Y-axis');

# 5. Save and show (Your save the whole figure)
ig.savefig('new.png')
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))
```

```python
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 10))
ax[0, 0].plot()
ax[0, 1].scatter()
```

### Which one should you use? (pyplot vs matplotlib oo method?)
- When plotting something quickly, it is okay to use the pyplot method
- When plotting something more advance it is okay to use the OO method

```python
## OO method from the scratch
o, m = plt.subplots(figsize = (10, 6))

# plot data
scatter = m.scatter(x = over_40['age'],
                     y = over_40['chol'],
                     c = over_40['target'])

# Customize the plot
m.set(title = 'Heart Diseas and choloesterol levels',
      xlabel = 'Age',
      ylabel = 'Cholesterol');
# Add a legend
m.legend(*scatter.legend_elements(), title = 'Target');

# Add a horizontal line
m.axhline(over_40['chol'].mean(),
           linestyle = ':');
```

```python
# Opject oriented method
b, t  = plt.subplots(figsize = (10, 6))

# Plotting the data
scatter = t.scatter(x = over_40['age'],
                    y = over_40['thalach'],
                    c = over_40['target'])
# Customizing
t.set(title = 'Age and highest heart rate',
      xlabel = 'Age',
      ylabel = 'Highest heart rate'
     )
# Legend
t.legend(*scatter.legend_elements(), title = 'Targets')

# Adding line 
t.axhline(over_40['thalach'].mean(),
          linestyle = '-.');
t.axvline(over_40['age'].mean(),
         linestyle = 'dotted');
```

```python
# Subplot of chol age and thalach
s, (a, b) = plt.subplots(nrows = 2,
                        ncols = 1,
                        figsize = (10, 10),
                        sharex = True)
# add data to a
scatter = a.scatter(x = over_40['age'],
                    y = over_40['chol'],
                    c = over_40['target'])
#customizing a
a.set(title = 'Highest cholesterol level',
      ylabel = 'Cholesterol level')
# Adding  legend
a.legend(*scatter.legend_elements(), title = 'Marks')

# Adding lines
a.axhline(over_40['chol'].mean(),
          linestyle = '-.');
a.axvline(over_40['age'].mean(),
         linestyle = '--');

# Adding data to b
scatter = b.scatter(x = over_40['age'],
                    y = over_40['thalach'],
                    c = over_40['target'])
# Customizing b
b.set(title = 'Highest heart rate',
      xlabel = 'Age',
      ylabel = 'Heart rate')
# Adding legend for b
b.legend(*scatter.legend_elements(), title = 'Marks')

# Adding line to b
b.axhline(over_40['thalach'].mean(),
         linestyle = ':')
b.axvline(over_40['age'].mean(),
         linestyle = '-');
# add a title to the figure
s.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
