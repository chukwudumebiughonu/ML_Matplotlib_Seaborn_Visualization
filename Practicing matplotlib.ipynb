## Matplotlib example workflow
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data 
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,10)) # (width, height)

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "simple Plot",
       xlabel = 'X-axis',
       ylabel = 'Y-axis')

# 5. Save and show (Your save the whole figure)
fig.savefig('Sample-plot.png')
```

## Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10,10))
plt.plot();
```

```python
# Plot to each different axis
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.barh(list(gadgetStore.keys()), list(gadgetStore.values()));
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                       figsize = (10, 5))
# Plot each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].barh(list(gadgetStore.keys()), list(gadgetStore.values()));
ax[1, 1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, it is okay to use the pyplot method
* When plotting somethin more advanced it is okay to use the OO method

```python
heartDisease.head()
```

```python
over_50 = heartDisease[heartDisease['age'] > 50]
over_50
```

```python
over_150 = heartDisease[heartDisease['thalach'] > 150]
over_150
```

```python
len(over_150)
```

```python
over_150.plot(kind = 'scatter',
             x = 'thalach',
             y = 'chol',
             c = 'target');
```

```python
# Using the pyplot method
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target');
```

```python
# OO method mixed with pyplot
fix, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target', 
            ax = ax);
# ax.set_xlim([45, 100]);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'])
# Customize the plot
ax.set(title= 'Heart Disease and cholesterol levels',
      xlabel= 'Age',
      ylabel= 'Cholesterol');
# Add a  legend
ax.legend(*scatter.legend_elements(), title = 'Target' );

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
# ax.axvline(over_50['chol'].mean())
# ax.axvline(over_50['age'].mean())
```

```python
# Method OO matplotlib
fig, ax = plt.subplots(figsize = (10,6))

# ploting oo matplotlib from scratch
scatter = ax.scatter(x = over_50['age'],
                    y = over_50['thalach'],
                    c = over_50['target'])
# customizing 
ax.set(title = 'Age and highest heart rate',
      xlabel = 'Age',
      ylabel = 'Highest heart rate'
      )
# legend
ax.legend(*scatter.legend_elements(), title = 'Targets')
# Adding line
ax.axhline(over_50['thalach'].mean(),
          linestyle = '--');
ax.axvline(over_50['age'].mean(),
          linestyle = '--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) =plt.subplots(nrows =2,
                              ncols =1,
                              figsize = (10,10),
                             sharex = True)
# Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])

# Customize ax0
ax0.set(title = 'Heart Disease and cholesterol levels',
      ylabel = 'Cholesterol level')
# Add legend 
ax0.legend(*scatter.legend_elements(), title = 'Targets')

# Add mean line
ax0.axhline(over_50['chol'].mean(),
          linestyle = '-.');
ax0.axvline(over_50['age'].mean(),
           linestyle = '--')
# Add data to ax1
scatter=ax1.scatter(x = over_50['age'],
                  y = over_50['thalach'],
                  c = over_50['target'])
# Customize ax1
ax1.set(title = 'Heart Disease and Highest heart rate',
       xlabel = 'Age',
       ylabel = 'Highest heart rate')
# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Targets')

# Add meanline to ax1
ax1.axhline(y = over_50['thalach'].mean(),
           linestyle =  '-')
ax1.axvline(x = over_50['age'].mean(),
           linestyle = ':')
# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold' );
```
